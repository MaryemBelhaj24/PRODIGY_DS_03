```python
#Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from io import StringIO
from sklearn.tree import export_graphviz
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.tree import export_text
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
# Data preparation: Define the features and target 
features = ['age', 'job', 'marital', 'education', 'default', 'balance', 'housing', 'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays', 'previous', 'poutcome']
target = 'deposit'
```

```python
missing_columns = set(['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome']) - set(bank.columns)
print("Missing columns:", missing_columns)
```

```python
bank.columns = bank.columns.str.strip()
```

```python
#Data Preprocessing; dealing with categorical data
bank = pd.get_dummies(bank, columns=['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'poutcome'], drop_first=True)
```

```python
#split data:
X = bank.drop(target, axis=1)
y = bank[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train
```

```python
X_train
X_test
```

```python
y_train
```

```python
y_test
```

```python
# Decision Tree Classifier
clf = DecisionTreeClassifier(random_state=42)
```

```python
clf.fit(X_train, y_train)
```

```python
# Predictions
y_pred = clf.predict(X_test)
# model evaluation
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python

conf_matrix = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", conf_matrix)
```

```python
class_report = classification_report(y_test, y_pred)
print("Classification Report:\n", class_report)
```

```python
confusion = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(confusion, annot=True, fmt="d", cmap="Blues")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
tree_structure = export_text(clf, feature_names=list(X.columns))
print("Decision Tree Structure:")
print(tree_structure)
```
